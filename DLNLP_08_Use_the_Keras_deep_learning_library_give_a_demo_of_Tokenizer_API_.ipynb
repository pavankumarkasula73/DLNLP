```python
# Importing the necessary libraries
!pip install tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer # Changed import statement to tensorflow.keras

# Sample text data
texts = [
    "Kalki 2898 AD is an upcoming Indian science fiction film.",
    "The movie is directed by Nag Ashwin and produced by Vyjayanthi Movies.",
    "The cast includes Prabhas, Deepika Padukone, and Amitabh Bachchan.",
    "The film is set in a futuristic world and promises groundbreaking visual effects.",
    "Kalki 2898 AD is one of the most anticipated movies of the year.",
    "The plot revolves around a dystopian future with advanced technology."
]

# Initialize the Tokenizer
tokenizer = Tokenizer(num_words=50)  # 'num_words' limits the vocabulary size to 50 most common words

# Fit the tokenizer on the text data
tokenizer.fit_on_texts(texts)

# Convert the texts to sequences of integers
sequences = tokenizer.texts_to_sequences(texts)

# Convert the texts to a matrix representation (binary, count, tf-idf, or frequency-based)
word_matrix = tokenizer.texts_to_matrix(texts, mode='tfidf')

# Get the word index (dictionary mapping words to their index)
word_index = tokenizer.word_index

# Display the results
print("Word Index:\n", word_index)
print("\nSequences:\n", sequences)
print("\nTF-IDF Word Matrix:\n", word_matrix)
```
