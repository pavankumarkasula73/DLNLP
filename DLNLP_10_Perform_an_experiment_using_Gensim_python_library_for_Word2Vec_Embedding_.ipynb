```python
# Importing necessary libraries
import gensim
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 1. Prepare the corpus
# Sample corpus: Each sentence is a list of words (tokenized text)
corpus = [
    ['Kalki', '2898', 'AD', 'is', 'an', 'upcoming', 'Indian', 'science', 'fiction', 'film'],
    ['The', 'movie', 'is', 'directed', 'by', 'Nag', 'Ashwin', 'and', 'produced', 'by', 'Vyjayanthi', 'Movies'],
    ['The', 'cast', 'includes', 'Prabhas', 'Deepika', 'Padukone', 'and', 'Amitabh', 'Bachchan'],
    ['The', 'film', 'is', 'set', 'in', 'a', 'futuristic', 'world', 'and', 'promises', 'groundbreaking', 'visual', 'effects'],
    ['Kalki', '2898', 'AD', 'is', 'one', 'of', 'the', 'most', 'anticipated', 'movies', 'of', 'the', 'year'],
    ['The', 'plot', 'revolves', 'around', 'a', 'dystopian', 'future', 'with', 'advanced', 'technology']
]

# 2. Train the Word2Vec model
# Initialize and train the Word2Vec model
word2vec_model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=1, workers=4)

# Summarize the model
print("Vocabulary size:", len(word2vec_model.wv.key_to_index))

# 3. Retrieve word vectors and display embeddings
# Retrieve the vocabulary
words = list(word2vec_model.wv.key_to_index)

# Extract the word vectors
word_vectors = word2vec_model.wv[words]

# 4. Visualize the word embeddings using PCA (Dimensionality Reduction)
# Use PCA to reduce the dimensionality to 2D for visualization
pca = PCA(n_components=2)
word_vectors_2d = pca.fit_transform(word_vectors)

# Plot the words in a 2D space
plt.figure(figsize=(10, 8))
plt.scatter(word_vectors_2d[:, 0], word_vectors_2d[:, 1])

# Annotate each point with the corresponding word
for i, word in enumerate(words):
    plt.annotate(word, xy=(word_vectors_2d[i, 0], word_vectors_2d[i, 1]))

plt.title("Word2Vec Embeddings (PCA Visualization)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```
